```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Q1.csv')

epochs = df['Epoch']
o_tacc = df['o_acc']
o_vacc = df['o_valacc']

ex1_tacc = df['Ex1_trainacc']
ex1_vacc = df['Ex1_valacc']

ex2_tacc = df['Ex2_trainacc']
ex2_vacc = df['Ex2_valacc']

fig, ax = plt.subplots()
fig.patch.set_facecolor('lightgrey')
ax.set_facecolor('lightgrey')

plt.plot(epochs,o_tacc , label='Given', marker='o', color='navy')
plt.plot(epochs,ex1_tacc , label='Ex1', marker='s', color='darkblue')
plt.plot(epochs,ex2_tacc , label='Ex2', marker='^', color='mediumblue')

for i, txt in enumerate(o_tacc):
    plt.annotate(f'{txt:.3f}', (epochs[i], o_tacc[i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8,color='navy')

for i, txt in enumerate(ex1_tacc):
    plt.annotate(f'{txt:.3f}', (epochs[i], ex1_tacc[i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8,color='darkblue')

for i, txt in enumerate(ex2_tacc):
    plt.annotate(f'{txt:.3f}', (epochs[i], ex2_tacc[i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8,color='mediumblue')

plt.title('Comparsion of Training Accuracy for Question 1')
plt.xlabel('Epochs')
plt.ylabel('Training Accuracy')
plt.legend()
plt.xticks(epochs)

plt.show()
```

```python
plt.figure(figsize=(12, 8))

fig, ax = plt.subplots()
fig.patch.set_facecolor('lightgrey')
ax.set_facecolor('lightgrey')

plt.plot(epochs,o_vacc , label='Given', marker='o', color='darkred')
plt.plot(epochs,ex1_vacc , label='Ex1', marker='s', color='red')
plt.plot(epochs,ex2_vacc , label='Ex2', marker='^', color='firebrick')

for i, txt in enumerate(o_vacc):
    plt.annotate(f'{txt:.3f}', (epochs[i], o_vacc[i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8,color='darkred')

for i, txt in enumerate(ex1_vacc):
    plt.annotate(f'{txt:.3f}', (epochs[i], ex1_vacc[i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8,color='red')

for i, txt in enumerate(ex2_vacc):
    plt.annotate(f'{txt:.3f}', (epochs[i], ex2_vacc[i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8,color='firebrick')

plt.title('Comparsion of Validation Accuracy for Question 1')
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')
plt.legend()
plt.xticks(epochs)

plt.show()
```
